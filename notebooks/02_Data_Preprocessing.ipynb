Import Libraries

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
```

```python
# Download the stopwords
nltk.download("stopwords")
```

Load Dataset

```python
df = pd.read_csv("../data/structured_it_resume_dataset.csv")
```

Combine Text Columns

```python
df["combined_text"] = (
    df["education"] + " " +
    df["Technical skills"] + " " +
    df["Soft skills"] + " " +
    df["Tools & frameworks"] + " " +
    df["Years of experience"].astype(str)+ " " +
    df["resume_paragraph"]
)

df[["job_role", "combined_text"]].head()
```

Text Cleaning

Lowercase,
Remove punctuation,
Remove special characters,
Remove stopwords,

```python
stop_words = set(stopwords.words("english"))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

Apply Cleaning

```python
df["clean_text"] = df["combined_text"].apply(clean_text)

df[["combined_text", "clean_text"]].head()
```

```python
df.to_csv("../data/processed_random_imbalanced.csv", index=False)
```
