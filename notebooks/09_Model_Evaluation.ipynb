Create Resuilt Table

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

results = [
    {"Model": "Logistic Regression", "Accuracy": 0.98, "Macro_F1": 0.97},
    {"Model": "Decision Tree", "Accuracy": 0.14, "Macro_F1": 0.14},
    {"Model": "Random Forest", "Accuracy": 0.95, "Macro_F1": 0.95},
    {"Model": "Gradient Boosting", "Accuracy": 0.96, "Macro_F1": 0.96},
    {"Model": "SVM", "Accuracy": 0.98, "Macro_F1": 0.98},
]

results_df = pd.DataFrame(results)
results_df
```

Plot Accuracy Comparison

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Model", y="Accuracy", data=results_df)
plt.xticks(rotation=45)
plt.title("Model Accuracy Comparison")
plt.ylim(0, 1)
plt.show()
```

Plot F1 Comparison

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Model", y="Macro_F1", data=results_df)
plt.xticks(rotation=45)
plt.title("Model Macro F1 Comparison")
plt.ylim(0, 1)
plt.show()
```

Select Best Model

```python
best_model = results_df.sort_values(by="Macro_F1", ascending=False).iloc[0]
print("Best Model Selected:")
print(best_model)
```
